```python
import pandas as pd
```

```python
# Lista de arquivos CSV
arquivos = ['movies_metadata.csv', 'features_r.csv', 'keywords.csv', 'links_small.csv', 'links.csv', 
            'ratings_small.csv', 'ratings.csv', 'credits.csv']

# Carregar os DataFrames
dfs = [pd.read_csv(arquivo, low_memory=False) for arquivo in arquivos]
```

```python
meta, fea, key, links_small, links, ratings_small, ratings, credits = dfs
```

```python
# Definir a largura máxima de colunas
pd.set_option('display.max_columns', 200)

# Listas de DataFrames e seus respectivos nomes
dfs = [meta, fea, key, links_small, links, ratings_small, ratings, credits]
names = ["Metadata", "Features", "Keywords", "Links Small", "Links", "Ratings Small", "Ratings", "Credits"]

# Exibir as duas primeiras linhas de cada DataFrame
for df, name in zip(dfs, names):
    print(name + ":")
    display(df.head(2))
    print("\n")
```

```python
# Lista de tuplas contendo o nome do DataFrame e o próprio DataFrame
dataframes = [("Metadata", meta), ("Features", fea), ("Keywords", key), ("Links Small", links_small),
              ("Links", links), ("Ratings Small", ratings_small), ("Ratings", ratings), ("Credits", credits)]

# Iterar sobre as tuplas e exibir o nome do DataFrame, sua forma e suas colunas
_ = [print(f"{name} shape: {df.shape} \n{name} columns: {df.columns.tolist()}\n") for name, df in dataframes]
```

```python
# Verificar NaNs nas colunas 'id' dos DataFrames antes da conversão
for name, df in [("Metadata", meta), ("Credits", credits), ("Keywords", key)]:
    print(f"NaNs in {name}['id']: ", df['id'].isna().sum())
```

```python
# Remover linhas onde 'id' em meta é NaN
meta.dropna(subset=['id'], inplace=True)
```

```python
# Converter coluna 'id' para string e depois para int64
meta['id'] = meta['id'].astype(str).str.extract('(\d+)')
credits['id'] = credits['id'].astype(str).str.extract('(\d+)')
key['id'] = key['id'].astype(str).str.extract('(\d+)')

# Mesclar os DataFrames
merged_data = pd.merge(pd.merge(meta, credits, on='id'), key, on='id')

merged_data.head(2)
```

```python
merged_data.columns = merged_data.columns.str.strip()
merged_data = merged_data.drop(['belongs_to_collection', 'homepage', 'imdb_id', 'poster_path', \
                                'video', 'title', 'status'], axis=1)
```

```python
# Convertendo a coluna 'id' para int64
merged_data['id'] = merged_data['id'].astype('int64')

# Verificando se há linhas com o mesmo valor de 'id'
duplicated_ids = merged_data[merged_data.duplicated(subset=['id'], keep=False)]
if not duplicated_ids.empty:
    print("Linhas com o mesmo valor de 'id':")
    print(duplicated_ids)
else:
    print("Não há linhas com o mesmo valor de 'id'.")
```

```python
# Excluir linhas duplicadas com base na coluna 'id'
merged_data.drop_duplicates(subset=['id'], keep='first', inplace=True)

# Verificar se ainda há linhas com o mesmo valor de 'id' (deve imprimir "Não há linhas com o mesmo valor de 'id'.")
duplicated_ids = merged_data[merged_data.duplicated(subset=['id'], keep=False)]
if not duplicated_ids.empty:
    print("Linhas com o mesmo valor de 'id':")
    print(duplicated_ids)
else:
    print("Não há linhas com o mesmo valor de 'id'.")
```

```python
# Importar a biblioteca re
import re

# Definir a função para extrair os gêneros de uma string JSON
def extrair_generos(string):
    return re.findall(r"'name': '(\w+)'", string)

# Aplicar a função para extrair os gêneros de cada linha da coluna 'genres'
merged_data['gêneros'] = merged_data['gêneros'].apply(extrair_generos)

# Exibir os resultados
print(merged_data['gêneros'])
```

```python
# Converter para datetime
data['data_de_lançamento'] = pd.to_datetime(data['data_de_lançamento'], errors='coerce')

# Dividir a coluna 'data_de_lançamento' em ano, mês e dia e converter para int
data['ano_lançamento'] = data['data_de_lançamento'].dt.year.astype(int)
data['mês_lançamento'] = data['data_de_lançamento'].dt.month.astype(int)
data['dia_lançamento'] = data['data_de_lançamento'].dt.day.astype(int)

# Preencher valores nulos na coluna 'data_de_lançamento' com NaT (Not a Time)
data['data_de_lançamento'] = data['data_de_lançamento'].fillna(pd.NaT)

# Criar coluna 'dia_semana_lançamento' com o dia da semana correspondente
data['dia_semana_lançamento'] = data['data_de_lançamento'].dt.day_name()

# Excluir a coluna 'data_de_lançamento'
data.drop(columns=['data_de_lançamento'], inplace=True)

data.head(1)

```

```python
# Dicionário para mapear os nomes dos dias da semana em português
dias_da_semana = {
    'Monday': 'Segunda-feira',
    'Tuesday': 'Terça-feira',
    'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira',
    'Friday': 'Sexta-feira',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo'
}

# Aplicar a tradução dos nomes dos dias da semana
data['dia_semana_lançamento'] = data['dia_semana_lançamento'].map(dias_da_semana)
data[['ano_lançamento', 'mês_lançamento', 'dia_lançamento','dia_semana_lançamento']].head(5)
```

```python
# Encontrar filmes duplicados e contar a frequência de cada título
filmes_duplicados = data[data.duplicated(subset=['título_original'], keep=False)]
contagem_filmes_duplicados = filmes_duplicados.groupby('título_original').size()

# Filtrar apenas os filmes que têm mais de uma ocorrência
filmes_duplicados = contagem_filmes_duplicados[contagem_filmes_duplicados > 1]

# Criar DataFrame com títulos duplicados e suas contagens
df_filmes_duplicados = pd.DataFrame({'Filme': filmes_duplicados.index, 'Repetições': filmes_duplicados.values})

# Ordenar o DataFrame em ordem decrescente de repetições
df_filmes_duplicados = df_filmes_duplicados.sort_values(by='Repetições', ascending=False)

# Resetar o índice do DataFrame
df_filmes_duplicados.reset_index(drop=True, inplace=True)

# Exibir o DataFrame resultante
df_filmes_duplicados.head(10)

```

```python
# Encontrar filmes duplicados e contar a frequência de cada combinação de título, diretor e ano de lançamento
filmes_duplicados = data[data.duplicated(subset=['título_original', 'diretor', 'elenco'], keep=False)]
contagem_filmes_duplicados = filmes_duplicados.groupby(['título_original', 'diretor', 'elenco']).size()

# Filtrar apenas os filmes que têm mais de uma ocorrência
filmes_duplicados = contagem_filmes_duplicados[contagem_filmes_duplicados > 1]

# Criar DataFrame com títulos duplicados e suas contagens
df_filmes_duplicados2 = pd.DataFrame({'Título Original': filmes_duplicados.index.get_level_values('título_original'),
                                     'Diretor': filmes_duplicados.index.get_level_values('diretor'),
                                     'Elenco': filmes_duplicados.index.get_level_values('elenco'),
                                     'Repetições': filmes_duplicados.values})

# Ordenar o DataFrame em ordem decrescente de repetições
df_filmes_duplicados2 = df_filmes_duplicados2.sort_values(by='Repetições', ascending=False)

# Resetar o índice do DataFrame
df_filmes_duplicados2.reset_index(drop=True, inplace=True)

# Exibir o DataFrame resultante
df_filmes_duplicados2.head(50)
```
